# Pelatihan Model Linear

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression, SGDRegressor
```

```python
# Dataset awal (dengan koefisien 3)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

plt.scatter(X, y)
plt.xlabel("X")
plt.ylabel("y")
plt.title("Dataset Linear Awal")
plt.show()
```

```python
# Latihan 1: Ubah koefisien menjadi 5
X1 = 2 * np.random.rand(100, 1)
y1 = 4 + 5 * X1 + np.random.randn(100, 1)

model1 = LinearRegression()
model1.fit(X1, y1)

print("Intercept:", model1.intercept_)
print("Koefisien:", model1.coef_)

X_new = np.array([[0], [2]])
y_pred1 = model1.predict(X_new)

plt.plot(X_new, y_pred1, "r-", label="Regresi Linear (Koef 5)")
plt.scatter(X1, y1, label="Data")
plt.legend()
plt.title("Regresi dengan Koefisien 5")
plt.show()
```

```python
# Latihan 2: Tambahkan outlier
X2 = np.append(X1, [[1.5], [1.8]], axis=0)
y2 = np.append(y1, [[30], [35]], axis=0)

model2 = LinearRegression()
model2.fit(X2, y2)

print("Intercept:", model2.intercept_)
print("Koefisien:", model2.coef_)

y_pred2 = model2.predict(X_new)

plt.plot(X_new, y_pred2, "r-", label="Regresi dengan Outlier")
plt.scatter(X2, y2, label="Data dengan Outlier")
plt.legend()
plt.title("Pengaruh Outlier pada Regresi Linear")
plt.show()
```

```python
# Latihan 3: Gunakan SGDRegressor
sgd = SGDRegressor(max_iter=1000, tol=1e-3, penalty=None, eta0=0.1)
sgd.fit(X1, y1.ravel())

print("Intercept (SGD):", sgd.intercept_)
print("Koefisien (SGD):", sgd.coef_)

y_sgd = sgd.predict(X_new)

plt.plot(X_new, y_sgd, "g-", label="SGDRegressor")
plt.scatter(X1, y1, label="Data")
plt.legend()
plt.title("Regresi dengan SGDRegressor")
plt.show()
```

## Kesimpulan

- Mengubah koefisien pada data akan mempengaruhi kemiringan garis regresi.
- Outlier dapat menggeser garis regresi secara signifikan.
- `SGDRegressor` memberikan hasil yang mendekati `LinearRegression` jika parameter dan data sesuai, tapi lebih sensitif terhadap data dan parameter awal.
